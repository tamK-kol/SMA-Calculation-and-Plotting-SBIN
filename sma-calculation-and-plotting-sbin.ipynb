```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Fetch historical stock data for SBIN
data = yf.download("SBIN.BO", start="2023-11-01", end="2023-11-30")
```

```python
# Calculated the 20-day Simple Moving Average (SMA)
data['SMA_20'] = data['Close'].rolling(window=20).mean()
```

```python
# Print the first few rows of the data with SMA for inspection
print(data.head())
```

```python
# Plot the closing prices and SMA
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Closing Prices')
plt.plot(data['SMA_20'], label='20=Day SMA', color='orange')
plt.title('SMA Stock Prices with 20-Day SMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
